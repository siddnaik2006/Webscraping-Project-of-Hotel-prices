```python
HotelName = soup.findAll('div',{'class':'f6431b446c a15b38c233'})
HotelName
```

```python
HotelName_list = []
for h in HotelName:
    HotelName_list.append(h.getText().strip())
    
HotelName_list
```

```python
Location = soup.findAll('span',{'class':'aee5343fdb def9bc142a'})
Location
```
